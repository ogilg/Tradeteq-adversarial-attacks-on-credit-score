### Features overview

All features all observable at classification time (T0) except for <b>isfailed</b> which is not revealed until T1. The target model is to predict <b>isfailed</b> given all T0 features.

Features <b>FieldXXXX</b> are values of certain accounting fields reported by company. Most companies do not report any accounting data. For each feature of this type, <b>hasFXXXX</b> contains the missing indicator.

Features <b>imd</b>, <b>imdu</b>, <b>lat</b>, <b>long</b>, <b>oa11</b>, <b>oac1</b>, <b>oac11</b>, <b>oac2</b>, <b>oseast1m</b>, <b>osnrth1m</b>, and <b>ru11ind</b> are all derived from the registration postcode <b>pcd</b>
